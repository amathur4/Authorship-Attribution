```python
import re
st = re.sub(r'[^\w\s]','',st)
st[:100]
```

```python
st = st.lower()
from collections import Counter
word_lst = Counter(st.split()).most_common()
word_lst
```

```python
#norm = ''
max_freq = -1
for (w,f) in word_lst:
    if max_freq < f:
        max_freq = f
#max_freq
norm = [(float(i)/max_freq) for (w,i) in word_lst]
norm1 = [(w,(float(i)/max_freq)) for (w,i) in word_lst]
#for (w,f) in word_lst:
#    norm += w + ' ' + str(float(f)/max_freq)
norm1
```

```python
norm_class = []
for (w,f) in norm1:
    if f>0 and f<=0.001:
        c = 19
    elif f>0.001 and f<=0.002:
        c = 18
    elif f>0.002 and f<=0.003:
        c = 17
    elif f>0.003 and f<=0.004:
        c = 16
    elif f>0.004 and f<=0.005:
        c = 15
    elif f>0.005 and f<=0.006:
        c = 14
    elif f>0.006 and f<=0.007:
        c = 13
    elif f>0.007 and f<=0.008:
        c = 12
    elif f>0.008 and f<=0.009:
        c = 11
    elif f>0.009 and f<=0.01:
        c = 10
    elif f>0.01 and f<=0.02:
        c = 9
    elif f>0.02 and f<=0.03:
        c = 8
    elif f>0.03 and f<=0.04:
        c = 7
    elif f>0.04 and f<=0.05:
        c = 6
    elif f>0.05 and f<=0.06:
        c = 5
    elif f>0.06 and f<=0.07:
        c = 4
    elif f>0.07 and f<=0.08:
        c = 3
    elif f>0.08 and f<=0.09:
        c = 2
    elif f>0.09 and f<=0.1:
        c = 1
    else:
        c = 0
    norm_class += [(w,c)]
norm_class
```

```python
avg_wrd_freq = open("avg_wrd_freq.txt","w")
for w in norm_class:
    #print(w + ' ' + str(c))
    avg_wrd_freq.write(' '.join(str(s) for s in w) + '\n')
avg_wrd_freq.close()
```
